```python
def bipolar_activation(x):
    if (x > 0):
        return 1
    elif (x < 0):
        return -1 
    else:
        return 0
```

```python
def unipolar_activation(x):
    if(x >= 0):
        return 1 
    else:
        return 0

def calculate_net(w1, x1, w2, x2, b):
    return (w1*x1 + w2*x2 + b)
```

```python
def perceptron_learning(x1, x2, d, w1, w2, b, activation, learning_rate=1):
    epochs = 0
    converged = False

    while not converged:
        error_count = 0
        print(f"\nFor Epoch {epochs + 1}: (w1 = {w1}, w2 = {w2}, b = {b})")
        print("x1\t x2\t d\t net\t o\t del_w1\t del_w2\t del_b\t new_w1\t new_w2\t new_b")

        for i in range(len(d)):
            net = calculate_net(w1, x1[i], w2, x2[i], b)
            o = activation(net)
            error = d[i] - o

            del_w1 = learning_rate * error * x1[i]
            del_w2 = learning_rate * error * x2[i]
            del_b = learning_rate * error

            new_w1 = w1 + del_w1
            new_w2 = w2 + del_w2
            new_b = b + del_b

            print(f"{x1[i]}\t {x2[i]}\t {d[i]}\t {net}\t {o}\t {del_w1}\t {del_w2}\t {del_b}\t {new_w1}\t {new_w2}\t {new_b}")

            if error != 0:
                w1, w2, b = new_w1, new_w2, new_b
                error_count += 1

        epochs += 1
        if error_count == 0:
            converged = True

    print("\nTraining complete")
    print(f"Final weights: w1 = {w1}, w2 = {w2}, bias = {b}")
    print(f"Total epochs: {epochs}")
    return w1, w2, b
```

```python
# AND gate with Bipolar activation 
bipolar_x1 = [-1, 1, -1, 1]
bipolar_x2 = [-1, -1, 1, 1]
bipolar_d = [-1, -1, -1, 1]

# AND gate with Unipolar activation 
unipolar_x1 = [0, 1, 0, 1]
unipolar_x2 = [0, 0, 1, 1]
unipolar_d = [0, 0, 0, 1]

w1, w2, b = 0, 0, 0

print('AND gate with Bipolar activation\n')
final_w1, final_w2, final_b = perceptron_learning(bipolar_x1, bipolar_x2, bipolar_d, w1, w2, b, bipolar_activation)

print('\nAND gate with Unipolar activation\n')
final_w1, final_w2, final_b = perceptron_learning(unipolar_x1, unipolar_x2, unipolar_d, w1, w2, b, unipolar_activation)
```

```python
# OR gate with Bipolar activation 
bipolar_x1 = [-1, 1, -1, 1]
bipolar_x2 = [-1, -1, 1, 1]
bipolar_d = [-1, 1, 1, 1]

# OR gate with Unipolar activation 
unipolar_x1 = [0, 1, 0, 1]
unipolar_x2 = [0, 0, 1, 1]
unipolar_d = [0, 1, 1, 1]

w1, w2, b = 0, 0, 0

print('OR gate with Bipolar activation\n')
final_w1, final_w2, final_b = perceptron_learning(bipolar_x1, bipolar_x2, bipolar_d, w1, w2, b, bipolar_activation)

print('\nOR gate with Unipolar activation\n')
final_w1, final_w2, final_b = perceptron_learning(unipolar_x1, unipolar_x2, unipolar_d, w1, w2, b, unipolar_activation)
```
